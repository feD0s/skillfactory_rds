# Подготовка датасета
В данном ноутбуке мы преобразуем ранее выгруженные с cian json-файл в датасет, с которым будем работать дальше

```python
import json
import pandas as pd
```

Откроем 2 ранее выгруженных json-файла и сделаем из них один.

```python
import os

with open("data_alup1.json", "r", encoding="utf8") as f:
    data = json.load(f)
df = pd.DataFrame(data)

for filename in os.listdir('data'):
    if filename.endswith('.json'):
        with open(os.path.join('data', filename)) as f:
            data = json.load(f)
        df2 = pd.DataFrame(data)
        df = df.append(df2, sort=True, ignore_index=True)
        print('appended ',filename)
```

Посмотрим на получившийся датасет. Также нам поможет JSON Viewer: http://jsonviewer.stack.hu/

```python
df.info()
df.head()
```
